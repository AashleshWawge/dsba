### Perform the following operations using Python on Movie data sets
### a. Create data subsets for different languages(Original Language).
### b. Merge two subsets
### c. Sort Data on Page total likes
### d. Transposing Data
### e. Melting Data to long format
### f. Casting data to wide format

```python
import pandas as pd
```

```python
df = pd.read_csv("tmdb_5000_movies.csv")
df.head()
```

###  Create data subsets for type of post

```python
english_movies = df[df['original_language'] == 'en']
french_movies = df[df['original_language'] == 'fr']
```

```python
print("\nSubset: English Movies")
english_movies.head()
```

```python
print("\nSubset: French Movies")
french_movies.head()
```

### Merge two subsets

```python
merged_df = pd.concat([english_movies, french_movies]) #merging two subsets
print("\nMerged Subset (Status + Photo):")
merged_df.head()
```

### Sort Data on Page total likes

```python
sorted_df = df.sort_values(by='popularity', ascending=False) #Sort Data on Page total likes
print("\nSorted Data by Page Total Likes:")
sorted_df.head()
```

### Transposing Data

```python
# Transposing means converting rows to columns and columns to rows.

transposed_df = df.transpose() #transposing the data
print("\nTransposed Data:")
transposed_df.head()
```

### Melting Data to long format

```python
# Melting is the process of turning wide-format data (where each variable has its own column) into long-format data
# (where there are only two or three columns: one for the variable name and one for the value).

melted_df = pd.melt(merged_df,
                    id_vars=['original_language', 'title'],
                    value_vars=['budget', 'popularity', 'revenue'])  # Melting Data to long format
print("\nMelted Data (Long Format):")
melted_df.head()
```

### Casting data to wide format

```python
# Casting involves transforming long-format data into a wide format, where each unique value in one column becomes its own separate column.

# Casting involves transforming long-format data into a wide format, where each unique value in one column becomes its own separate column.

pivot_df = melted_df.pivot_table(values='value', index='original_language', columns='variable', aggfunc='mean')
print("\nPivoted (Wide Format):")
pivot_df.head()
```
