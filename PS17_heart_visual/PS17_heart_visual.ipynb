## Visualize the Heart disease dataset by plotting the following graphs using Python. (Define objective for every graph)
### a. Histograms
### b. Dot Plots
### c. Bar Plots
### d. Line Charts
### e. Add Histogram and Scatter plot to box plot.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('heart.csv')
df.head()
```

## HISTOGRAM
### Objective: To understand the distribution of individual numerical features like age, chol (cholesterol), thalach (max heart rate).

```python
# Plot histograms for important features
df[['age', 'chol', 'thalach']].hist(bins=15, figsize=(12, 6), color='skyblue', edgecolor='black')
plt.suptitle('Histograms of Age, Cholesterol, and Max Heart Rate', fontsize=15)
plt.show()
```

## DOT PLOTS
### Objective: To visualize the frequency of categorical features like cp (chest pain type) or sex.

```python
# Dot plot (strip plot) for chest pain type vs target
plt.figure(figsize=(8, 5))
sns.stripplot(x='cp', y='age', data=df, jitter=True)
plt.title('Dot Plot of Chest Pain Type vs Age')
plt.xlabel('Chest Pain Type')
plt.ylabel('Age')
plt.show()
```

## BAR PLOTS
### Objective: To compare the mean cholesterol levels across different types of chest pain (cp) or across gender (sex).

```python
plt.figure(figsize=(8, 5))
sns.barplot(x='cp', y='chol', hue='cp', data=df, errorbar='sd', palette='pastel', legend=False)
plt.title('Average Cholesterol Levels per Chest Pain Type')
plt.xlabel('Chest Pain Type')
plt.ylabel('Cholesterol')
plt.show()
```

## LINE CHARTS
### Objective: To observe trend patterns, for example how maximum heart rate (thalach) varies with age.

```python
# Line chart: average max heart rate across ages
df_age_thalach = df.groupby('age')['thalach'].mean()

plt.figure(figsize=(10, 6))
df_age_thalach.plot(kind='line', marker='o', color='green')
plt.title('Line Chart of Age vs Average Max Heart Rate')
plt.xlabel('Age')
plt.ylabel('Average Max Heart Rate')
plt.grid(True)
plt.show()
```

## Add Histogram and Scatter plot to box plot.
### Objective: To show detailed distribution (histogram and scatter) along with outliers (box plot) for variables like age or chol.

```python
# Create combined boxplot, histogram and scatter (strip) for Age
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Box plot + Scatter
sns.boxplot(x='target', y='age', hue='cp', data=df, ax=axes[0], palette='cool')
sns.stripplot(x='target', y='age', data=df, ax=axes[0], color='red', jitter=True, alpha=0.5)
axes[0].set_title('Box Plot and Scatter of Age by Heart Disease')

# Histogram
sns.histplot(data=df, x='age', hue='target', multiple='stack', palette='cool', ax=axes[1])
axes[1].set_title('Histogram of Age by Heart Disease')

plt.show()
```
