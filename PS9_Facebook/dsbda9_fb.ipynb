### Perform the following operations using Python on the Facebook metrics data sets
### a. Create data subsets for type of post
### b. Merge two subsets
### c. Sort Data on Page total likes
### d. Transposing Data
### e. Melting Data to long format
### f. Casting data to wide format

```python
import pandas as pd 
```

```python
df = pd.read_csv('dataset_Facebook.csv', sep=';') #loading the dataset 
df.head()
```

###  Create data subsets for type of post

```python
status_df = df[df['Type'] == 'Status'] #creating a subset status
photo_df = df[df['Type'] == 'Photo']  #creating a subset photo
```

```python
print("\nSubset: Status Posts")
status_df.head()
```

```python
print("\nSubset: Photo Posts")
photo_df.head()
```

### Merge two subsets

```python
merged_df = pd.concat([status_df, photo_df]) #merging two subsets
print("\nMerged Subset (Status + Photo):")
merged_df.head()
```

### Sort Data on Page total likes

```python
sorted_df = df.sort_values(by='Page total likes', ascending=False) #Sort Data on Page total likes
print("\nSorted Data by Page Total Likes:")
print(sorted_df[['Page total likes', 'Type']].head())
```

### Transposing Data

```python
# Transposing means converting rows to columns and columns to rows.

transposed_df = df.transpose() #transposing the data
print("\nTransposed Data:")
transposed_df.head()
```

### Melting Data to long format

```python
# Melting is the process of turning wide-format data (where each variable has its own column) into long-format data 
# (where there are only two or three columns: one for the variable name and one for the value).

melted_df = pd.melt(df,
                    id_vars=['Type', 'Category'],
                    value_vars=['Page total likes', 'Post Month', 'Post Weekday'])  #Melting Data to long format 
print("\nMelted Data (Long Format):")
melted_df.head()
```

### Casting data to wide format

```python
# Casting involves transforming long-format data into a wide format, where each unique value in one column becomes its own separate column.

pivot_df = df.pivot_table(values='Page total likes', index='Category', columns='Type', aggfunc='mean')
print("\nPivoted (Wide Format):")
pivot_df.head()
```
