### b. Pie Charts
### c. Box Plots
### d. Scatter Plots

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load Adult dataset without header
df = pd.read_csv('adult.csv', header=None)
```

```python
# Manually assign column names
column_names = [
    'age', 'workclass', 'fnlwgt', 'education', 'education_num',
    'marital_status', 'occupation', 'relationship', 'race', 'sex',
    'capital_gain', 'capital_loss', 'hours_per_week', 'native_country', 'income'
]

df.columns = column_names

# Check
df.head()
```

```python
df.isnull().sum()
```

```python
df.shape
```

## HISTOGRAM
### Objective: To understand the distribution of individual numerical features like age, chol (cholesterol), thalach (max heart rate).

```python
plt.figure(figsize=(8, 5))
df['age'].hist(bins=20, color='skyblue', edgecolor='black')
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Count")
plt.grid(False)
plt.show()
```

## PIE CHARTS
### Objective: To visualize the proportion of people with and without heart disease (target variable).

```python
# Pie chart for gender distribution
plt.figure(figsize=(6,6))
gender_counts = df['sex'].value_counts()
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=140, colors=['#ff9999','#66b3ff'])
plt.title('Gender Distribution in Dataset')
plt.axis('equal')
plt.show()
```

## BOX PLOTS
### Objective: To observe the spread, median, and outliers in features like cholesterol or age.

```python
# Box plot: hours per week vs income
plt.figure(figsize=(8,6))
sns.boxplot(x='income', y='hours_per_week', hue='income', data=df, palette='Set2')
plt.title('Working Hours by Income Category')
plt.xlabel('Income')
plt.ylabel('Hours per Week')
plt.show()
```

## SCATTER PLOTS
### Objective: To see the relationship between two continuous variables  e.g., age vs thalach (max heart rate).

```python
# Scatter plot: age vs hours_per_week
plt.figure(figsize=(8,6))
sns.scatterplot(x='age', y='hours_per_week', data=df, hue='sex', alpha=0.5)
plt.title('Age vs Hours per Week by Gender')
plt.xlabel('Age')
plt.ylabel('Hours per Week')
plt.legend(title='Gender')
plt.show()
```

## Add boxplots to a scatterplot
### Objective: To simultaneously show relationship between two variables and their individual distributions (boxplots on margins).

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='income', y='hours_per_week', hue='income', data=df, whis=1.5, palette='pastel')
sns.stripplot(x='income', y='hours_per_week', data=df, color='black', alpha=0.3, jitter=True)
plt.title("Boxplot + Scatter of Hours per Week by Income")
plt.xlabel("Income")
plt.ylabel("Hours per Week")
plt.grid(True)
plt.show()
```
