```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('heart.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
df.shape
```

## HISTOGRAM
### Objective: To understand the distribution of individual numerical features like age, chol (cholesterol), thalach (max heart rate).

```python
# Plot histograms for important features
df[['age', 'chol', 'thalach']].hist(bins=15, figsize=(12, 6), color='skyblue', edgecolor='black')
plt.suptitle('Histograms of Age, Cholesterol, and Max Heart Rate', fontsize=16)
plt.show()
```

## PIE CHARTS
### Objective: To visualize the proportion of people with and without heart disease (target variable).

```python
# Pie chart for target variable
heart_disease_counts = df['target'].value_counts()

plt.figure(figsize=(6,6))
plt.pie(heart_disease_counts, labels=['No Disease', 'Disease'], autopct='%1.1f%%', colors=['lightcoral', 'lightgreen'], startangle=90)
plt.title('Heart Disease Proportion')
plt.show()
```

## BOX PLOTS
### Objective: To observe the spread, median, and outliers in features like cholesterol or age.

```python
# Box plot for Cholesterol levels
plt.figure(figsize=(8,5))
sns.boxplot(y='chol', data=df, color='lightblue')
plt.title('Boxplot of Cholesterol')
plt.ylabel('Cholesterol')
plt.show()
```

## SCATTER PLOTS
### Objective: To see the relationship between two continuous variables — e.g., age vs thalach (max heart rate).

```python
# Scatter plot for Age vs Max Heart Rate
plt.figure(figsize=(8,5))
sns.scatterplot(x='age', y='thalach', hue='target', data=df, palette='cool')
plt.title('Age vs Max Heart Rate')
plt.xlabel('Age')
plt.ylabel('Max Heart Rate')
plt.legend(title='Heart Disease')
plt.show()
```

## Add boxplots to a scatterplot
### Objective: To simultaneously show relationship between two variables and their individual distributions (boxplots on margins).

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='target', y='thalach', hue = data=df, palette='Set2')
sns.stripplot(x='target', y='thalach', data=df, jitter=True, color='black', alpha=0.5)
plt.title("Max Heart Rate vs Heart Disease (Box + Scatter)")
plt.xlabel("Heart Disease (0 = No, 1 = Yes)")
plt.ylabel("Max Heart Rate")
plt.grid(True)
plt.show()
```
